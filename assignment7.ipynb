1. What is a RESTful API?

A RESTful API is an application programming interface that follows the principles of REST (Representational State Transfer). It enables communication between a client and a server over HTTP using standard methods like GET, POST, PUT, DELETE, and PATCH. Each resource is identified using a unique URL, and data is typically exchanged in JSON format.

Example: A RESTful API endpoint like /users/123 with a GET method fetches the user details with ID 123.

2. Explain the concept of API specification

An API specification is a detailed description of how an API works. It outlines available endpoints, request types, parameters, expected responses, error messages, authentication methods, and data formats. It acts as a contract between the API provider and consumer, ensuring clear communication and integration.

Example: The OpenAPI (formerly Swagger) specification allows developers to define API behavior in a standard format and auto-generate documentation or client SDKs.

3. What is Flask, and why is it popular for building APIs?

Flask is a lightweight Python web framework that enables easy and quick development of web applications and REST APIs. It is popular because of its simplicity, flexibility, and minimal setup. Flask supports extensions for form validation, database integration, authentication, and more.

Example: Developers often use Flask to build microservices or REST APIs with minimal code and overhead.

4. What is routing in Flask?

Routing in Flask is the mechanism that maps URLs to specific Python functions (called view functions). It helps define what response should be given when a user accesses a particular URL.

Example: The route /home can be connected to a function that returns a homepage HTML or JSON response.

5. How do you create a simple Flask application?

To create a simple Flask application, you import Flask, create an instance of it, define routes using decorators, and run the app using app.run().

Example: A basic Flask app may return "Hello, World!" on accessing the root URL /.

6. What are HTTP methods used in RESTful APIs?

RESTful APIs use the following HTTP methods:

GET: Retrieve data from the server

POST: Send new data to the server

PUT: Update existing data

DELETE: Remove data

PATCH: Apply partial updates to a resource

Example: A POST request to /users can create a new user entry in the database.

7. What is the purpose of the @app.route() decorator in Flask?

The @app.route() decorator in Flask is used to bind a URL to a view function. It tells Flask which function to execute when a specific route is accessed.

Example: @app.route('/about') will direct the /about URL to the corresponding function.

8. What is the difference between GET and POST HTTP methods?

GET requests are used to fetch data and are idempotent (do not change server state).

POST requests are used to submit data and typically change the server's state (e.g., create a new record).

Example: A GET request to /product/1 fetches product data, while a POST request to /product adds a new product.

9. How do you handle errors in Flask APIs?

Flask allows you to handle errors using custom error handlers and the @app.errorhandler() decorator. You can return custom error messages and status codes.

Example: A 404 error page can be returned when a user tries to access a non-existing route.

10. How do you connect Flask to a SQL database?

You can connect Flask to a SQL database using the Flask-SQLAlchemy extension. You configure the database URI and create models to interact with the database.

Example: Connecting to a SQLite database using SQLAlchemy can be done by setting SQLALCHEMY_DATABASE_URI = 'sqlite:///data.db'.

11. What is the role of Flask-SQLAlchemy?

Flask-SQLAlchemy integrates SQLAlchemy with Flask, providing a high-level ORM (Object Relational Mapper) that allows Python classes to be mapped to database tables. It simplifies database operations like querying, updating, and deleting records.

12. What are Flask blueprints, and how are they useful?

Blueprints in Flask allow modular organization of applications by grouping related views, templates, and static files into reusable components. This promotes cleaner code and easier maintenance in large applications.

13. What is the purpose of Flask's request object?

The request object in Flask gives access to incoming request data such as form data, query parameters, headers, and JSON payloads.

Example: You can retrieve form data from a POST request using request.form['name'].

14. How do you create a RESTful API endpoint using Flask?

You define a route with @app.route() and specify HTTP methods. Inside the view function, you handle the logic for that endpoint.

Example: A GET method on /user/<id> returns user information based on the ID.

15. What is the purpose of Flask's jsonify() function?

The jsonify() function converts Python dictionaries or lists into JSON responses, making it easy to return structured data from Flask APIs.

16. Explain Flask’s url_for() function

The url_for() function dynamically generates URLs using the name of the view function. It helps avoid hardcoding URLs and supports query parameters.

17. How does Flask handle static files (CSS, JavaScript, etc.)?

Flask serves static files from the static/ folder. These files can be accessed using url_for('static', filename='file.css') inside templates.

18. What is an API specification, and how does it help in building a Flask API?

An API specification documents how your API works, what inputs and outputs it expects, and how to handle errors. It ensures consistency across development and integration.

Example: Tools like Swagger generate interactive documentation from API specs, helping frontend and backend teams collaborate.

19. What are HTTP status codes, and why are they important in a Flask API?

HTTP status codes indicate the outcome of a request:

200 OK: Successful request

404 Not Found: Resource doesn't exist

500 Internal Server Error: Unexpected server error

They help clients understand the result of their request and handle responses appropriately.

20. How do you handle POST requests in Flask?

You specify the method as POST in @app.route() and access submitted data using request.form, request.json, or request.get_json().

21. How would you secure a Flask API?

Use HTTPS to encrypt data

Authenticate users using tokens (e.g., JWT)

Validate and sanitize user inputs

Implement rate limiting and logging

Protect sensitive endpoints with role-based access control

22. What is the significance of the Flask-RESTful extension?

Flask-RESTful is an extension that adds support for quickly building REST APIs. It simplifies routing, serialization, request parsing, and integrates well with Flask-SQLAlchemy.

23. What is the role of Flask’s session object?

The session object stores user data between HTTP requests. It allows you to persist data like login state using secure cookies.
