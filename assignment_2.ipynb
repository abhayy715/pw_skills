1. What is a random variable in probability theory?


A random variable is a numerical outcome of a random experiment. It assigns a real number to each possible outcome in the sample space. For example, when tossing a coin, if heads = 1 and tails = 0, the result becomes a random variable.

2. What are the types of random variables?


Discrete Random Variable: Takes distinct, countable values like 0,1,2... (e.g., number of cars passing a toll booth).

Continuous Random Variable: Takes any value in an interval, including fractions (e.g., height, temperature).

3. What is the difference between discrete and continuous distributions?


Discrete Distribution: Deals with countable outcomes (like number of calls received).

Continuous Distribution: Deals with measurable quantities that can take an infinite number of values within a range (like weight or time).

4. What are probability distribution functions (PDF)?


A PDF for a continuous variable shows the relative likelihood of a random variable taking on a given value. The area under the PDF curve within an interval represents the probability of the variable falling in that range.

5. How do cumulative distribution functions (CDF) differ from PDF?


The CDF gives the probability that a random variable is less than or equal to a certain value. Unlike PDF, which gives the density at a specific point, the CDF provides a cumulative probability from the lower limit up to that point.

6. What is a discrete uniform distribution?


A discrete uniform distribution assigns equal probability to all possible discrete outcomes. An example is rolling a fair die—each face (1 to 6) has the same probability of 1/6.

7. What are the key properties of a Bernoulli distribution?


Only two possible outcomes: success (1) or failure (0).

One trial.

Mean = p; Variance = p(1-p).
Example: Tossing a coin once (head or tail).

8. What is the binomial distribution, and how is it used in probability?


It represents the number of successes in a fixed number of independent Bernoulli trials. It is used to model scenarios like "how many heads will appear in 10 coin tosses."

9. What is the Poisson distribution and where is it applied?


The Poisson distribution measures the probability of a number of events happening in a fixed period or space if these events occur independently at a constant average rate. Example: Number of emails received per hour.

10. What is a continuous uniform distribution?


A continuous uniform distribution means every value in a given range is equally likely to occur. For example, if you randomly select a time between 1 and 2 PM, every moment in that hour has the same probability.

11. What are the characteristics of a normal distribution?


Symmetrical, bell-shaped curve.

Defined by mean (center) and standard deviation (spread).

Most values lie within ±3 standard deviations from the mean.

12. What is the standard normal distribution, and why is it important?


A normal distribution with a mean of 0 and standard deviation of 1. It simplifies calculation of probabilities and is used to standardize scores (Z-scores).

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?


CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size grows, regardless of the population's distribution. This allows statisticians to make inferences about the population.

14. How does the Central Limit Theorem relate to the normal distribution?


CLT ensures that the distribution of the sample mean becomes approximately normal as the sample size increases, allowing the use of the normal distribution in hypothesis testing even when the population is not normal.

15. What is the application of Z statistics in hypothesis testing?


Z-statistics are used to determine how far a sample statistic is from the population mean, measured in standard deviations. It helps decide whether to reject the null hypothesis in large samples.

16. How do you calculate a Z-score, and what does it represent?


Z = (X – μ) / σ
It tells how many standard deviations an individual data point (X) is from the population mean (μ). A Z-score helps in standardizing data for comparison.

17. What are point estimates and interval estimates in statistics?


Point Estimate: A single best guess (like sample mean) of a population parameter.

Interval Estimate: A range (like confidence interval) likely to contain the population parameter with a certain confidence level.

18. What is the significance of confidence intervals in statistical analysis?


A confidence interval provides a range within which the true population parameter is likely to lie. It reflects the uncertainty or variability in the estimate from a sample.

19. What is the relationship between a Z-score and a confidence interval?


Z-scores define the margins of error for confidence intervals. For example, a 95% confidence level corresponds to Z = ±1.96, setting the boundaries for the confidence interval around the estimate.

20. How are Z-scores used to compare different distributions?


Z-scores standardize different data sets by converting them to a common scale (standard normal), allowing meaningful comparison between different distributions or variables.

21. What are the assumptions for applying the Central Limit Theorem?


Samples must be independent.

The sample size should be large (n ≥ 30 is preferred).

Population variance should be finite.

Data should be identically distributed.

22. What is the concept of expected value in a probability distribution?


The expected value is the theoretical mean of a random variable, representing the average outcome if the experiment is repeated many times. It is calculated by summing the products of each value and its probability.

23. How does a probability distribution relate to the expected outcome of a random variable?


The probability distribution determines the expected value by assigning probabilities to all possible outcomes. It reflects the long-term average result if the experiment is repeated infinitely.
