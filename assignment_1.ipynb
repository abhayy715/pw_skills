## THEORY QUESTIONS

1. What is statistics, and why is it important?

    Statistics is the science of collecting, analyzing, interpreting, and presenting data in a meaningful way. It helps us understand patterns, make predictions, and make informed decisions. It is widely used in research, business, medicine, economics, and social sciences to solve real-world problems based on data.

2. What are the two main types of statistics?

    The two main types are Descriptive and Inferential Statistics. Descriptive statistics summarize and describe data using tools like mean, median, and charts. Inferential statistics use sample data to make generalizations about a population through estimation and hypothesis testing.

3. What are descriptive statistics?

    Descriptive statistics are methods for summarizing raw data into a more understandable form. They include measures like mean, median, mode, standard deviation, and graphs like histograms or boxplots. These help in quickly understanding trends and patterns within a dataset.

4. What is inferential statistics?

    Inferential statistics allow us to make predictions or inferences about a population based on a sample. It uses probability theory and methods like confidence intervals, hypothesis testing, and regression analysis to draw conclusions beyond the observed data.

5. What is sampling in statistics?

    Sampling is the process of selecting a subset of individuals from a larger population to estimate characteristics of the whole group. It’s used when studying an entire population is impractical, and it helps save time, effort, and cost while still giving meaningful results.

6. What are different sampling methods?

    Common sampling methods include:

    Random sampling – each individual has an equal chance.

    Stratified sampling – population divided into groups and samples taken from each.

    Systematic sampling – every k-th member is chosen.

    Cluster sampling – entire groups are randomly selected.


7. What is the difference between random and non-random sampling?

    In random sampling, every unit in the population has an equal and known chance of being selected, which reduces bias. Non-random sampling relies on human judgment or convenience, which may lead to biased and unrepresentative results.

8. Define and give examples of qualitative and quantitative data.

    Qualitative data refers to descriptive information, like colors, names, or categories (e.g., gender, city names).

    Quantitative data is numerical and can be measured, such as height, weight, or test scores.

9. What are the different types of data in statistics?

    There are four main data types:

    Nominal: Categorized by names and there is no order (e.g., gender).

    Ordinal: Ordered categories (e.g., ranks).

    Interval: Numeric but no true zero (e.g., temperature).

    Ratio: Numeric with a true zero (e.g., weight).

10. Explain nominal, ordinal, interval, and ratio levels of measurement.

    Nominal: Labels without order (e.g., fruits: apple, banana).

    Ordinal: Categories with order but no fixed spacing (e.g., satisfaction: high, medium, low).

    Interval: Ordered, equally spaced, but no true zero (e.g., dates).

    Ratio: Like interval, but has a meaningful zero (e.g., income, age).

11. What is the measure of central tendency?

    It’s a statistical measure that identifies the center of a dataset. The three main measures are mean (average), median (middle value), and mode (most frequent value). These help summarize large data into a single representative value.

12. Define mean, median, and mode.

    Mean is the average value, found by dividing the sum by the number of values and also called as arithmetic mid value.

    Median is the middle number when values are ordered and also known as physical mid value.

    Mode is the value that occurs most frequently in the dataset.

13. What is the significance of the measure of central tendency?

    It simplifies a dataset by giving a single value that best represents the data. It helps in comparing datasets, understanding trends, and making decisions in fields like business, education, and health.

14. What is variance, and how is it calculated?

    Variance measures how far each data point is from the mean. It’s calculated by taking the average of squared differences from the mean. A high variance means data is widely spread; low variance means it's tightly grouped.

15. What is standard deviation, and why is it important?

    Standard deviation is the square root of variance and indicates the spread of data in the same unit as the original data. It is important because it helps understand how consistent or varied the values in a dataset are.

16. Define and explain the term range in statistics.

    Range is the simplest measure of dispersion and is found by subtracting the smallest value from the largest value in the dataset. It gives a quick idea of how spread out the data is.

17. What is the difference between variance and standard deviation?

    Variance is the average of squared differences from the mean, and its unit is the square of the original unit. Standard deviation is its square root and is expressed in the same unit as the data, making it easier to interpret.

18. What is skewness in a dataset?

    Skewness measures the asymmetry of a dataset's distribution. A symmetrical distribution has zero skewness. Positive skew has a longer right tail, and negative skew has a longer left tail.

19. What does it mean if a dataset is positively or negatively skewed?

    A positively skewed dataset has a longer tail on the right side, indicating most values are low. A negatively skewed dataset has a longer tail on the left, meaning most values are high.

20. Define and explain kurtosis.

    Kurtosis measures the “tailedness” of a distribution. High kurtosis indicates heavy tails and more extreme values (outliers), while low kurtosis suggests light tails and fewer outliers.

21. What is the purpose of covariance?

    Covariance indicates whether two variables change together. A positive value means they increase together, while a negative value shows an inverse relationship. It helps understand how variables are related.

22. What does correlation measure in statistics?

    Correlation measures both the strength and direction of a linear relationship between two variables. Values range from -1 to 1, where 1 means perfect positive, -1 perfect negative, and 0 no correlation.

23. What is the difference between covariance and correlation?

    Both show relationships, but covariance gives the direction without a fixed scale, while correlation standardizes the relationship, making it easier to compare across different datasets.

24. What are some real-world applications of statistics?

    Statistics is used in healthcare (disease prediction), business (sales forecasting), sports (performance tracking), education (test analysis), and government (census and policy decisions).
