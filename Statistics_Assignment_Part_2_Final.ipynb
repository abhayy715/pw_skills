
## Theory Questions

1. **Hypothesis Testing:** Method to check assumptions about population parameters using sample data.

2. **Null vs Alternative Hypothesis:**  
- H0: No effect/difference.  
- H1: Effect/difference exists.

3. **Significance Level:**  
Threshold probability for Type I error. Common values: 0.01, 0.05, 0.1.

4. **P-value:**  
Probability of obtaining test results as extreme as the observed, under H0.

5. **Interpretation of P-value:**  
- P ≤ α: Reject H0.  
- P > α: Fail to reject H0.

6. **Type I and II Errors:**  
- Type I (α): False positive.  
- Type II (β): False negative.

7. **One-tailed vs Two-tailed Tests:**  
- One-tailed: Directional.  
- Two-tailed: Non-directional.

8. **Z-test:**  
For large samples (n>30) with known variance.

9. **Z-score:**  
Z = (x̄ - μ) / (σ / √n)

10. **T-distribution:**  
For small samples (n<30), unknown variance.

11. **Z-test vs T-test:**  
- Z-test: Large n, known σ.  
- T-test: Small n, unknown σ.

12. **T-test:**  
Compare sample mean to population mean or two sample means.

13. **Relation:**  
Both compare means; T adjusts for small samples.

14. **Confidence Interval:**  
Range estimate for population parameter.

15. **Margin of Error:**  
MoE = Z * (σ / √n)

16. **Bayes' Theorem:**  
P(A|B) = (P(B|A) * P(A)) / P(B)

17. **Chi-square Distribution:**  
Used in categorical variable testing.

18. **Chi-square Goodness of Fit:**  
Compares observed to expected frequencies.

19. **F-distribution:**  
Used in ANOVA, variance tests.

20. **ANOVA Test:**  
Compares means across groups.  
Assumptions: Independence, normality, equal variances.

21. **ANOVA Types:**  
One-way, Two-way, Repeated measures.

22. **F-test:**  
Compares variances between groups.


practical questions

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.proportion import proportions_ztest

print("1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.")
data = np.random.normal(loc=50, scale=10, size=100)
def z_test(data, pop_mean, pop_std):
    sample_mean = np.mean(data)
    n = len(data)
    z = (sample_mean - pop_mean) / (pop_std/np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p
z_score, p_value = z_test(data, 50, 10)
print("Z-Test: Z-Score =", z_score, "P-Value =", p_value, "\n")

print("2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.")
sample = np.random.normal(55, 10, 100)
t_stat, p_val = stats.ttest_1samp(sample, 50)
print("One-sample T-Test: T-Stat =", t_stat, "P-Value =", p_val, "\n")

print("3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.")
print("One-sample Z-Test Z-score:", z_test(data, 50, 10)[0], "\n")

print("4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.")
def plot_z_test(z_score):
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, stats.norm.pdf(x))
    plt.axvline(x=z_score, color='red', linestyle='--')
    plt.axvline(x=-z_score, color='red', linestyle='--')
    plt.title('Two-tailed Z-Test Decision Region')
    plt.show()
plot_z_test(z_score)
print("\n")

print("5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.")
def plot_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, stats.norm.pdf(x), label='H0')
    plt.fill_between(x, 0, stats.norm.pdf(x), where=(x > stats.norm.ppf(1-alpha)), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, stats.norm.pdf(x, loc=2), where=(x < stats.norm.ppf(beta)), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title('Type I and II Errors')
    plt.show()
plot_errors()
print("\n")

print("6. Write a Python program to perform an independent T-test and interpret the results.")
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
t_stat, p_val = stats.ttest_ind(group1, group2)
print("Independent T-Test: T-Stat =", t_stat, "P-Value =", p_val, "\n")

print("7. Perform a paired sample T-test using Python and visualize the comparison results.")
before = np.random.normal(50, 5, 30)
after = before + np.random.normal(1, 2, 30)
t_stat, p_val = stats.ttest_rel(before, after)
print("Paired T-Test: T-Stat =", t_stat, "P-Value =", p_val, "\n")

print("8. Simulate data and perform both Z-test and T-test, then compare the results using Python.")
print("Z-Test Result:", z_test(data, 50, 10))
print("T-Test Result:", stats.ttest_1samp(data, 50), "\n")

print("9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.")
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return interval
print("Confidence Interval:", confidence_interval(data), "\n")

print("10. Write a Python program to calculate the margin of error for a given confidence level using sample data.")
def margin_of_error(std, n, z=1.96):
    return z * (std / np.sqrt(n))
print("Margin of Error:", margin_of_error(10, 100), "\n")

print("11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.")
def bayes_theorem(p_a, p_b_given_a, p_b):
    return (p_b_given_a * p_a) / p_b
print("Bayes Result:", bayes_theorem(0.5, 0.7, 0.6), "\n")

print("12. Perform a Chi-square test for independence between two categorical variables in Python.")
obs = np.array([[10, 20], [20, 40]])
chi2, p, dof, ex = stats.chi2_contingency(obs)
print("Chi-square Test: Chi2 =", chi2, "P-Value =", p, "\n")

print("13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.")
print("Expected Frequencies:\n", ex, "\n")

print("14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.")
obs = [50, 30, 20]
exp = [40, 40, 20]
chi2, p = stats.chisquare(obs, f_exp=exp)
print("Goodness of Fit Test: Chi2 =", chi2, "P-Value =", p, "\n")

print("15. Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics.")
x = np.linspace(0, 10, 1000)
plt.plot(x, stats.chi2.pdf(x, df=2))
plt.title('Chi-square Distribution')
plt.show()
print("\n")

print("16. Implement an F-test using Python to compare the variances of two random samples.")
group1 = np.random.normal(10, 2, 20)
group2 = np.random.normal(10, 3, 20)
f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
print("F-Test Statistic:", f_stat, "\n")

print("17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.")
group1 = np.random.normal(20, 5, 30)
group2 = np.random.normal(22, 5, 30)
group3 = np.random.normal(21, 5, 30)
f_stat, p_val = stats.f_oneway(group1, group2, group3)
print("One-way ANOVA: F-Stat =", f_stat, "P-Value =", p_val, "\n")

print("18. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.")
print("Shapiro Test for Normality:", stats.shapiro(group1))
print("Levene’s Test for Equal Variance:", stats.levene(group1, group2, group3), "\n")

print("19. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.")
df = pd.DataFrame({
    'score': np.random.randn(60),
    'factor1': np.repeat(['A','B'], 30),
    'factor2': np.tile(['X','Y','Z'], 20)
})
model = ols('score ~ C(factor1) + C(factor2) + C(factor1):C(factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-way ANOVA Table:\n", anova_table, "\n")

print("20. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.")
x = np.linspace(0, 5, 1000)
plt.plot(x, stats.f.pdf(x, dfn=5, dfd=2))
plt.title('F-distribution')
plt.show()
print("\n")

print("21. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.")
sns.boxplot(data=[group1, group2, group3])
plt.title('One-way ANOVA Boxplot')
plt.show()
print("\n")

print("22. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.")
data = np.random.normal(50, 10, 100)
print("T-Test:", stats.ttest_1samp(data, 50), "\n")

print("23. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.")
obs_var = np.var(data, ddof=1)
chi2_stat = (len(data)-1)*obs_var/100
print("Chi-square for variance:", chi2_stat, "\n")

print("24. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.")
count = np.array([30, 45])
nobs = np.array([100, 120])
z_stat, p_val = proportions_ztest(count, nobs)
print("Z-test for Proportions: Z-Stat =", z_stat, "P-Value =", p_val, "\n")

print("25. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.")
sns.histplot(np.random.f(5, 2, 1000), kde=True)
plt.title('F-distribution Plot')
plt.show()
print("\n")

print("26. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.")
sim_data = np.random.randint(1, 4, 100)
freq = np.bincount(sim_data)[1:]
chi2, p = stats.chisquare(freq)
print("Chi-square Goodness of Fit: Chi2 =", chi2, "P-Value =", p)
```
